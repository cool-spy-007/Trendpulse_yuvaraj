```python
"""
Task 4 - Visualization
Create charts from the analyzed data
"""

import pandas as pd
import matplotlib.pyplot as plt
import os

# Create outputs folder
if not os.path.exists('outputs'):
    os.makedirs('outputs')

# Load data
df = pd.read_csv('data/trends_analysed.csv')
print(f"Loaded {len(df)} rows")

# Chart 1 - Top 10 stories by score
print("Creating Chart 1...")
top10 = df.nlargest(10, 'score')

# Shorten long titles
titles = []
for title in top10['title']:
    if len(title) > 50:
        titles.append(title[:47] + '...')
    else:
        titles.append(title)

plt.figure(figsize=(10, 6))
plt.barh(range(len(top10)), top10['score'], color='steelblue')
plt.yticks(range(len(top10)), titles)
plt.xlabel('Score')
plt.ylabel('Story Title')
plt.title('Top 10 Stories by Score')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig('outputs/chart1_top_stories.png', dpi=150)
plt.close()

# Chart 2 - Stories per category
print("Creating Chart 2...")
cat_counts = df['category'].value_counts().sort_index()

colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A', '#98D8C8']

plt.figure(figsize=(10, 6))
bars = plt.bar(cat_counts.index, cat_counts.values, color=colors)
plt.xlabel('Category')
plt.ylabel('Number of Stories')
plt.title('Stories per Category')
plt.xticks(rotation=45, ha='right')

# Add numbers on bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
            f'{int(height)}', ha='center', va='bottom')

plt.tight_layout()
plt.savefig('outputs/chart2_categories.png', dpi=150)
plt.close()

# Chart 3 - Scatter plot
print("Creating Chart 3...")
popular = df[df['is_popular'] == True]
not_popular = df[df['is_popular'] == False]

plt.figure(figsize=(10, 6))
plt.scatter(not_popular['score'], not_popular['num_comments'], 
           alpha=0.6, s=50, c='lightcoral', label='Not Popular')
plt.scatter(popular['score'], popular['num_comments'], 
           alpha=0.6, s=50, c='green', label='Popular')
plt.xlabel('Score')
plt.ylabel('Number of Comments')
plt.title('Score vs Comments (by Popularity)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('outputs/chart3_scatter.png', dpi=150)
plt.close()

# Bonus - Dashboard
print("Creating Dashboard...")
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('TrendPulse Dashboard', fontsize=16, fontweight='bold')

# Chart 1 in dashboard
ax1 = axes[0]
top10 = df.nlargest(10, 'score')
titles = []
for title in top10['title']:
    if len(title) > 40:
        titles.append(title[:37] + '...')
    else:
        titles.append(title)
ax1.barh(range(len(top10)), top10['score'], color='steelblue')
ax1.set_yticks(range(len(top10)))
ax1.set_yticklabels(titles, fontsize=8)
ax1.set_xlabel('Score')
ax1.set_title('Top 10 Stories')
ax1.invert_yaxis()

# Chart 2 in dashboard
ax2 = axes[1]
cat_counts = df['category'].value_counts().sort_index()
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A', '#98D8C8']
bars = ax2.bar(cat_counts.index, cat_counts.values, color=colors)
ax2.set_xlabel('Category')
ax2.set_ylabel('Count')
ax2.set_title('Stories per Category')
ax2.tick_params(axis='x', rotation=45, labelsize=8)
for bar in bars:
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height,
            f'{int(height)}', ha='center', va='bottom', fontsize=8)

# Chart 3 in dashboard
ax3 = axes[2]
popular = df[df['is_popular'] == True]
not_popular = df[df['is_popular'] == False]
ax3.scatter(not_popular['score'], not_popular['num_comments'], 
           alpha=0.6, s=40, c='lightcoral', label='Not Popular')
ax3.scatter(popular['score'], popular['num_comments'], 
           alpha=0.6, s=40, c='green', label='Popular')
ax3.set_xlabel('Score')
ax3.set_ylabel('Comments')
ax3.set_title('Score vs Comments')
ax3.legend(fontsize=8)
ax3.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('outputs/dashboard.png', dpi=150)
plt.close()

print("\nAll charts saved to outputs/ folder!")
print("Files created:")
print("  - chart1_top_stories.png")
print("  - chart2_categories.png")
print("  - chart3_scatter.png")
print("  - dashboard.png")
```
